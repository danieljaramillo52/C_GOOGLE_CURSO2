## **Ramas**

Linea de desarollo independiente. 

Puedes hacer pruebas o borrador. Existe una rama maestra y otras de mas tipo.


###  **git branch**

- Muestra las ramas, las crea, puedes cambiar entre ellas. 

```bash
$ git branch
* main
```

Solo tenemos por el momento la rama main. Podemos crear otra rama utilizando tambien 

```bash
git branch new-feature 
```

Donde ``new-feature`` es el nombre de la nueva rama. 

Obtenemos como resultado: 

```bash
ovejo@Daniel_Jara MINGW64 ~/Documents/Xpert-group/Cursos XpertGroup/Certificacion GOOGLE automatizacion python/C_GOOGLE_CURSOS (main)
$ git branch new-feature

ovejo@Daniel_Jara MINGW64 ~/Documents/Xpert-group/Cursos XpertGroup/Certificacion GOOGLE automatizacion python/C_GOOGLE_CURSOS (main)
$ git branch
* main
  new-feature
```

Podemos ver luego la nueva rama usando solo git branch

### **Navegar entre ramas**

- Podemos cambiar de rama utilizando el comando

**git switch nombre_rama_a_cambiar**

Por ejemplo pasemos a la nueva rama que creamos.

## **Fusión de Ramas**

Vamos a trabajar con una fusión simple. Ya habiamos creado la rama anterior. 

Podemos pasar lo hecho en otra rama (Siempre y cuando no genere conflictos) **Cuando hay cambios incompatibles entre las dos ramas, cuidado con la fusión de conflictos**

Con lo hecho en una rama a parte que tiene una nueva funcionalidad. 

Lo podemos realizar utilizando el comando **``git merge``**

1. Primero vamos a la rama principal

2. Realizamos **``git merge nombre_otra_rama``** para unir la otra_rama a la principal.


```bash 
ovejo@Daniel_Jara MINGW64 ~/Documents/Xpert-group/Cursos XpertGroup/Certificacion GOOGLE automatizacion python/C_GOOGLE_CURSOS (main)
$ git merge new-feature-two
Updating 8d6ef38..7536e56
Fast-forward
 free_memory.py | 6 ++++++
 1 file changed, 6 insertions(+)
 create mode 100644 free_memory.py
```

Podemos ver ahora que el archivo **free_memory.py** se añadió al flujo de trabajo de la rama princiapal, **la rama principal está al día**

Al hacer la fusión ``Ambas ramas apuntarán al mismo commit``
La fusión combina datos ramificados y el hisotrial. 

## **Fusión de conflictos**

### **Merge conflict** 

Ediciones en la misma parte del mismo archivo. 

Editemos el archivo **free_memory.py** tanto en la rama ``main`` como en la rama ``new-feature-two``

**Dentro de la rama main**

```python
#!/usr/bin/env python3

def main():
    print("Todo está Ok!")
main()
```

**Dentro de la rama new-features-two**

```python
#!/usr/bin/env python3

def main():
    """El método main encapsula el procedimiento de lógica del proyecto y lo lanza"""
main()
```

Podemos ver claramente que este cambio causará un conflicto. 

Luego de hacer en ambas ramas su respectivo **git commit -a -m "mensaje"**, vamos a intentar fusionarlas. **``Recordemos estar en la rama main``**


Podemos notar que tenemos un conflico

```bash
$ git merge new-feature-two
Auto-merging free_memory.py
CONFLICT (content): Merge conflict in free_memory.py
Automatic merge failed; fix conflicts and then commit the result.
```

git **status** (fragmento extraido que nos interesa) nos dice: 
```bash 
Unmerged paths:
  (use "git add <file>..." to mark resolution)
        both modified:   free_memory.py

```

Que tenemos dos archivos para merge. Por lo tanto debemos ``abortar la fusión```
O ``solucionar el conflicto``


Al entrar al archivo podemos ver algo como esto: 

```python
#!/usr/bin/env python3

<<<<<<< HEAD
def main():
    print("Todo está Ok!")
main()
=======

def main():
    """El método main encapsula el procedimiento de lógica del proyecto y lo lanza"""


main()
>>>>>>> new-feature-two
```
Podemos corregir directamente usar uno o ambos para que el conflicto desaparezca.


El editor de código Visual Studio no sa alternativas para elegir uno de los dos escensarios o hacer una nueva correción. Luego permite hacer un commit de todos los cambios. De manera bastante intuitiva.


### **Abortar un merge.** 

Podemos utilizar el comando **git merge --abort**


## **Git rebase**

Reescribe el historial de commits para colocar los cambios encima de otra base. 

Sobreescribe los cambios posibles desde el "main" a la nueva rama la rama. (Pone commit desde main a la otra rama)

```bash
git switch feature
git rebase main
```

**``Git NO toca main.``**
Lo que hace es:

Encuentra el último ancestro común entre main y feature y actualiza todo cambio de main en feature

Toma tus commits de feature (por ejemplo, D, E).

Antes:
main:    A -- B -- C
feature:            \-- D -- E

git rebase main  (en feature)

Después:
main:    A -- B -- C
feature:             \-- D′ -- E′   (re-jugados encima de C)


# Reaplica tu trabajo encima de main (solo afecta a new-feature)
git switch new-feature
git rebase main           # resuelve conflictos si aparecen: git add ..., git rebase --continue

Ahora podemos hacer un merge limpio... para actualizar el main con el contenido que teniamos en la otra rama (Sin commit de merge)  
# Ahora avanza main a la punta de new-feature sin merge commit
git switch main
git merge --ff-only new-feature
